* Parameters

```python
n = 10**9
nb_cycles = 7

alpha = 1.4
beta = 0.6
delta = 1.3
gamma = 2 # use gamma = None to make a simulation without density-dependence

birth_rates = [0.3, 0.45, 1, 1.1, 1.1]
death_rates = [0.1]*5
rates = [birth_rates, death_rates]
```

* Simulate a bottlenecked population

**n:** $n$ parameter (inverse of total mutation rate)  
**nb_cycles:** number of cycles to simulate  
**alpha:** parameter for final wild-type population size ($n^\alpha$), between 1 and 2  
**beta:** parameter for initial wild-type population size ($n^\beta$), between 0 and 1  
**delta:** parameter for the difference between high mutation rate ($\frac{1}{n}$) and low mutation rate ($\frac{1}{n^\delta}$), should be greater than 1  
**rates:** intrinsic birth and deaths rates of the 5 subpopulations (WT, 10 mutant, 01 mutant, double mutant from 10, double mutant from 01)  
**gamma:** parameter for carrying capacity ($K=n^\gamma$), set to None to make a simulation without density-dependence  
**sep_double:** whether or not to plot separately the two double mutants subpopulations (from 10 and from 01)  
**leg_loc:** legend location, set to None for no legend

```python
%%capture --no-stdout --no-display

plot_sim(n, nb_cycles, alpha, beta, delta, rates, gamma = gamma, sep_double = True, leg_loc = "upper right")
```

* Simulate several populations

**n_sim:** number of populations to simulate  
**n:** $n$ parameter (inverse of total mutation rate)  
**nb_cycles:** number of cycles to simulate  
**delta:** parameter for the difference between high mutation rate ($\frac{1}{n}$) and low mutation rate ($\frac{1}{n^\delta}$), should be greater than 1  
**rates:** intrinsic birth and deaths rates of the 5 subpopulations (WT, 10 mutant, 01 mutant, double mutant from 10, double mutant from 01)  
**gamma:** parameter for carrying capacity ($K=n^\gamma$), set to None to make a simulation without density-dependence

```python
%%capture

n_sim = 300
sim = many_sim_pop(n_sim, n, nb_cycles, delta, rates, gamma = gamma)
```

* Plot observed scenarios

**A:** list of $\alpha$ parameters  
**B:** list of $\beta$ parameters  
**S:** list of observed scenarios  
**predictions:** wheter or not to plot threshold lines for large $n$ predictions  
**delta:** value of $\delta$ parameter used for the simulations (used to plot predictions)  
**rates:** value of rates used for the simulations (used to plot predictions)

```python
make_fig_paths(sim[0], sim[1], sim[2], predictions = True, rates = rates, delta = delta)
```
